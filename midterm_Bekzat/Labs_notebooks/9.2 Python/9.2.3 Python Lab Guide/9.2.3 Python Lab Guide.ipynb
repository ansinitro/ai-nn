## 1.1 Regular Expressions
The examples extract emails and numeric identifiers from noisy text.


```python
import re

message = "student Sundetkhan Bekzat uses bekzat@example.com and ticket AI-2026-042"
email_pattern = re.compile(r"[\w.+-]+@[\w.-]+\.[A-Za-z]{2,}")
ticket_pattern = re.compile(r"AI-\d{4}-\d{3}")

print("email:", email_pattern.findall(message))
print("ticket:", ticket_pattern.search(message).group(0))
```

## 1.2 Validation Helpers
Compiled patterns make repeated validation clearer.


```python
import re

phone = re.compile(r"^\+?\d{10,15}$")
samples = ["+77011234567", "8700-111", "1234567890"]
for item in samples:
    print(item, bool(phone.match(item)))
```

## 1.3 Iterable Checks
The function avoids treating strings as collections to flatten.


```python
from collections.abc import Iterable

def flatten_once(values):
    result = []
    for item in values:
        if isinstance(item, Iterable) and not isinstance(item, (str, bytes)):
            result.extend(item)
        else:
            result.append(item)
    return result

print(flatten_once([[1, 2], "AI", (3, 4), 5]))
```

## 1.4 Decorators
A timing decorator keeps measurement logic separate from the function being tested.


```python
import time

def measured(func):
    def wrapped(*args, **kwargs):
        started = time.perf_counter()
        value = func(*args, **kwargs)
        elapsed_ms = (time.perf_counter() - started) * 1000
        print(f"{func.__name__} completed in {elapsed_ms:.3f} ms")
        return value
    return wrapped


@measured
def square_sum(limit):
    return sum(number * number for number in range(limit))


print(square_sum(1000))
```
