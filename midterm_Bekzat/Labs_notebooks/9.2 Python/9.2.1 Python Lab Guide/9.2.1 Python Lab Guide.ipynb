## 1.3 Branches and Loops
A deterministic grading function replaces interactive input so the notebook remains executable.


```python
def grade_label(score):
    if score >= 90:
        return "A"
    if score >= 75:
        return "B"
    if score >= 60:
        return "C"
    return "Needs practice"

for value in [95, 82, 68, 41]:
    print(f"score={value}: {grade_label(value)}")

row_count = 5
for left in range(1, row_count + 1):
    cells = [f"{left}x{right}={left * right}" for right in range(1, left + 1)]
    print(" | ".join(cells))
```

## 1.4 Functions
The Fibonacci task is implemented as a reusable function with validation.


```python
def fibonacci_terms(limit):
    if limit <= 0:
        return []
    values = [0, 1]
    while len(values) < limit:
        values.append(values[-1] + values[-2])
    return values[:limit]

print(fibonacci_terms(8))
```

## 1.5 Object-Oriented Programming
A small lab record class demonstrates attributes, methods, and class-level counters.


```python
class LabRecord:
    created = 0

    def __init__(self, owner, section):
        self.owner = owner
        self.section = section
        self.notes = []
        LabRecord.created += 1

    def add_note(self, note):
        self.notes.append(note)

    def summary(self):
        return f"{self.owner}: {self.section}, notes={len(self.notes)}"


record = LabRecord("Sundetkhan Bekzat", "9.2.1")
record.add_note("data types completed")
record.add_note("loops completed")
print(record.summary())
print("records created:", LabRecord.created)
```
